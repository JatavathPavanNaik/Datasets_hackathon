```python
import os  
import base64
from openai import AzureOpenAI  
from dotenv import load_dotenv


deployment = os.getenv("DEPLOYMENT_NAME", "gpt-4o")
import os


load_dotenv()
subscription_key = os.getenv("openai_api_key")
endpoint=os.getenv("openai_end_point")


# Initialize Azure OpenAI Service client with key-based authentication    
client = AzureOpenAI(  
    azure_endpoint=endpoint,  
    api_key=subscription_key,  
    api_version="2024-05-01-preview",
)
    
    
# IMAGE_PATH = "YOUR_IMAGE_PATH"
# encoded_image = base64.b64encode(open(IMAGE_PATH, 'rb').read()).decode('ascii')
text='''Today I feel really good about my progress at work. Everything seems to be going smoothly'''
user_name = session_data['Name']
mood = session_data['Mood']

    
# Include speech result if speech is enabled  
messages = [
            {"role": "system", 
             "content": '''
             Input descriptions:
                - **mood**: The new mood entered by the user.
                - **user_description**: A brief description from the user explaining the reason for their mood.

                Example input:
                - **mood**: Sad
                - **user_description**: "I feel down today, nothing seems to go right."

                Output:
                Provide a sentiment classification for the user's input. The output should be strictly one of the following: **Positive**, **Negative**, or **Neutral**.'''},
    {"role": "user", "content": f"User Info: {user_name}\nMood: {mood}\nQuery: {text}"}
]
    
# Generate the completion  
completion = client.chat.completions.create(  
    model=deployment,
    messages=messages,
    max_tokens=800,  
    temperature=0.7,  
    top_p=0.95,  
    frequency_penalty=0,  
    presence_penalty=0,
    stop=None,  
    stream=False
)

print(completion.to_json())  
```

```python
import json
import re
# Extract the content from the response
content = completion.to_json()

# Use regex to extract text between the **
pattern = r"(?<=\*\*)(.*?)(?=\*\*)"

# Apply the regex pattern to get the text between **
match = re.search(pattern, content)

# If a match is found, print it
if match:
    extracted_content = match.group(0)  # This will return "Positive"
    print(extracted_content)
else:
    print("No match found.")
```

```python
import pandas as pd
import uuid
import random
import faker

# Initialize Faker for generating fake names and user data
fake = faker.Faker()

# Function to generate a fake session and add to DataFrame
def add_fake_session_data(df, sentiment_output):
    # Generate fake session data
    fake_session_id = str(uuid.uuid4())  # Generate a unique fake session ID
    fake_name = fake.name()  # Fake name using Faker
    fake_mood = random.choice(['Sad', 'Angry', 'Happy', 'Neutral'])  # Randomly choose a mood
    fake_user_description = f"I am feeling {fake_mood.lower()} today."  # Basic description based on mood

    # Create new row with fake data
    new_row = {
        "Session_ID": fake_session_id,
        "Name": fake_name,
        "Mood": fake_mood,
        "Sentiment_Emotion": sentiment_output,  # Output sentiment classification (e.g., Positive/Negative/Neutral)
        "User_ID": f"ID_{random.randint(1, 100)}",  # Fake user ID
        "Mood_Sentiment_Ratio": {'Mood': {'Angry': 0.0, 'Sad': 0.0, 'Happy': 0.0}},  # Placeholder mood ratio
        "User_Description": fake_user_description,
        "Behavior_Description": f"User seems to be {fake_mood.lower()}."  # Simple behavior description
    }

    # Append the new row to the DataFrame using pd.concat()
    new_row_df = pd.DataFrame([new_row])  # Convert new_row dictionary to DataFrame
    df = pd.concat([df, new_row_df], ignore_index=True)  # Concatenate the new row to the existing DataFrame

    return df

# Example output sentiment from model (this would come from your sentiment analysis model)
sentiment_output = "Positive"  # For example, this is the output sentiment

# Add the new fake session data and update the sentiment
session_data = add_fake_session_data(session_data, sentiment_output)
```

```python
# Variables to keep track of mood and sentiment counts up to the current point
mood_counts = {"Angry": 0, "Sad": 0, "Happy": 0, "Neutral": 0, "Excited": 0}
sentiment_counts = {"Positive": 0, "Negative": 0, "Neutral": 0}

# Loop through the DataFrame and update the ratios dynamically
mood_sentiment_data = []

for i in range(len(session_data)):
    mood = session_data.at[i, 'Mood']
    sentiment = session_data.at[i, 'Sentiment_Emotion']
    
    # Update mood and sentiment counts
    mood_counts[mood] += 1
    sentiment_counts[sentiment] += 1
    
    # Calculate dynamic ratios for mood and sentiment
    mood_ratio = {key: mood_counts[key] / (i + 1) for key in mood_counts}
    sentiment_ratio = {key: sentiment_counts[key] / (i + 1) for key in sentiment_counts}
    
    # Combine both mood and sentiment ratios into a single entry
    mood_sentiment_data.append({
        'Mood_Sentiment_Ratio': {
            'Mood': mood_ratio,
            'Sentiment': sentiment_ratio
        }
    })

# Add the combined ratio data as a new column
session_data['Mood_Sentiment_Ratio'] = [entry['Mood_Sentiment_Ratio'] for entry in mood_sentiment_data]
```
