```python
import pandas as pd
import numpy as np
import scipy.stats as stats
from sklearn.preprocessing import LabelEncoder

# Function to calculate Cramér's V
def cramers_v(x, y):
    crosstab = pd.crosstab(x, y)
    chi2, p, dof, expected = stats.chi2_contingency(crosstab)
    n = crosstab.sum().sum()
    return np.sqrt(chi2 / (n * min(crosstab.shape) - 1))

# Convert categorical variables to numeric using LabelEncoder
encoder = LabelEncoder()
encoded_data = Life_Style.apply(encoder.fit_transform)

# Create an empty correlation matrix
cramers_v_matrix = pd.DataFrame(index=encoded_data.columns, columns=encoded_data.columns)

# Calculate Cramér's V for each pair of columns
for col1 in encoded_data.columns:
    for col2 in encoded_data.columns:
        cramers_v_matrix.loc[col1, col2] = cramers_v(encoded_data[col1], encoded_data[col2])

# Convert the values in the matrix to numeric types
cramers_v_matrix = cramers_v_matrix.astype(float)
```

```python
cramers_v_matrix
import seaborn as sns
sns.heatmap(cramers_v_matrix.corr())
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Set the threshold for strong correlation
threshold = 0.5

# Filter the Cramér's V matrix for correlations above the threshold
strong_correlations = cramers_v_matrix.where(cramers_v_matrix > threshold)

# Create a network graph
G = nx.Graph()

# Add edges for pairs of features with strong correlations
for col1 in strong_correlations.columns:
    for col2 in strong_correlations.index:
        if pd.notna(strong_correlations.loc[col2, col1]) and strong_correlations.loc[col2, col1] > threshold:
            G.add_edge(col1, col2, weight=strong_correlations.loc[col2, col1])

# Draw the network graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, k=0.15, iterations=20)  # Positioning for better visualization
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10, font_weight='bold', edge_color='gray')

# Display the plot
plt.title('Network Visualization of Features with Strong Cramér\'s V Correlations')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import seaborn as sns
from kmodes.kmodes import KModes
from tqdm import tqdm
```

```python
Life_Style = Life_Style.drop(columns='Customer ID').astype('category')
cost = []
K = range(1, 15)
data=Life_Style

for k in tqdm(K):
    km = KModes(n_clusters=k, init="Cao", n_init=10, verbose=1)
    km.fit_predict(data)
    cost.append(km.cost_)
```

```python
from kneed import KneeLocator

# Automatically find the elbow (optimal k)
knee_locator = KneeLocator(range(1, len(cost)+1), cost, curve="convex", direction="decreasing")
optimal_k = knee_locator.knee

print(f"Optimal number of clusters (k) suggested by KneeLocator: {optimal_k}")

# Plot with the detected elbow point
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cost)+1), cost, marker='o', linestyle='--', color='b')
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal k={optimal_k}')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 2️⃣: Choose optimal k (you can use an elbow method alternative)
optimal_k = 4  # Set manually or determine using cost function

# Step 3️⃣: Fit K-Modes with Optimal k
kmodes_final = KModes(n_clusters=optimal_k, init='Cao', n_init=10, random_state=42)
clusters = kmodes_final.fit_predict(data)  # Assign cluster labels

# Step 4️⃣: Visualize Cluster Distribution
plt.figure(figsize=(8, 6))
plt.hist(clusters, bins=range(optimal_k + 1), color='blue', alpha=0.7, edgecolor='black')
plt.xticks(range(optimal_k))
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.title(f'K-Modes Clustering (k={optimal_k}) Distribution')
plt.grid()
plt.show()

# Step 5️⃣: Add Cluster Labels to DataFrame
Life_Style['Cluster_KModes'] = clusters
```

# Kmeans and PCA

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import math
```

```python
# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(df_PCA[encoded_differently])
# Apply PCA
pca_full = PCA()
X_pca_full = pca_full.fit_transform(Hormones_Enzymes.drop(columns='Customer ID'))

# Explained Variance Analysis
explained_variance_ratio = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# Select Optimal Components (95% variance)
threshold = 0.95
n_components_optimal = np.argmax(cumulative_variance >= threshold) + 1
print(f"Optimal number of PCA components: {n_components_optimal}")

# Plot Explained Variance with Annotation
plt.figure(figsize=(8,5))
plt.plot(cumulative_variance, marker='o', linestyle='--', color='b', label='Cumulative Explained Variance')

# Marking the optimal component with an annotation
plt.axvline(x=n_components_optimal, linestyle='--', color='r', label=f'Optimal Components ({n_components_optimal})')
plt.scatter(n_components_optimal, cumulative_variance[n_components_optimal-1], color='red', s=100, zorder=3)  # Highlight the point

# Adding annotation text
plt.annotate(f'{n_components_optimal} Components\n{cumulative_variance[n_components_optimal-1]*100:.2f}% Variance',
             xy=(n_components_optimal, cumulative_variance[n_components_optimal-1]),
             xytext=(n_components_optimal + 2, cumulative_variance[n_components_optimal-1] - 0.05),
             arrowprops=dict(facecolor='red', shrink=0.05))

# Labels and Grid
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components')
plt.legend()
plt.grid()
plt.show()

# Apply PCA with optimal components
pca_final = PCA(n_components=n_components_optimal)
X_pca_final = pca_final.fit_transform(Hormones_Enzymes.drop(columns='Customer ID'))

# Display the reduced dataset shape
print(f"Original Shape: {Hormones_Enzymes.drop(columns='Customer ID').shape}, Reduced Shape: {X_pca_final.shape}")
```

```python
# Get PCA Loadings (Component Contributions)
pca_loadings = pd.DataFrame(
    pca_final.components_.T,
    index=Hormones_Enzymes.drop(columns='Customer ID').columns, 
    columns=[f'PC{i+1}' for i in range(pca_final.n_components_)]
)
```

```python
# Compute absolute PCA loadings
abs_pca_loadings = pca_loadings.abs()

# Count how many PCA components each feature contributes to (non-zero loadings)
feature_coverage = (abs_pca_loadings > 0).sum(axis=1)  # Count non-zero contributions

# Compute the total strength of each feature across all components
feature_strength = abs_pca_loadings.sum(axis=1)  # Sum of absolute loadings

# Normalize feature strength (relative importance)
normalized_strength = feature_strength / feature_strength.sum()

# Find the best PCA component for each feature (max strength PC)
best_pc = abs_pca_loadings.idxmax(axis=1)  # PC with highest contribution

# Create a DataFrame with the results
feature_analysis = pd.DataFrame({
    'Best_PC': best_pc,  # How many components a feature contributes to
    'Total_Strength': feature_strength,  # Overall impact across all components
    # 'Relative_Strength': normalized_strength  # Normalized strength (sum to 1)
})

display(feature_analysis)

plt.figure(figsize=(14, 10))  # Increase figure size for better spacing
ax = feature_analysis['Total_Strength'].sort_values(ascending=True).plot(kind='barh', color='royalblue', alpha=0.8)

# Adding annotations for best PCA component
for i, (index, row) in enumerate(feature_analysis.sort_values(by='Total_Strength', ascending=True).iterrows()):
    plt.text(row['Total_Strength'] + 0.05, i, row['Best_PC'], va='center', fontsize=10, fontweight='bold', color='black')

# Enhancing plot appearance
plt.title("Feature Contribution Strength Across All PCA Components", fontsize=14, fontweight='bold')
plt.xlabel("Total Contribution Strength", fontsize=12)
plt.ylabel("Features", fontsize=12)
plt.yticks(fontsize=10)  # Increase y-tick font size
plt.xticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Display improved horizontal bar plot
plt.show()
```

```python
# Use the Elbow Method to find the optimal number of clusters
wcss = []  # Within-Cluster Sum of Squares (WCSS)

for k in range(1, 20):  # Trying k values from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca_final)
    wcss.append(kmeans.inertia_)
```

```python
from kneed import KneeLocator

# Automatically find the elbow (optimal k)
knee_locator = KneeLocator(range(1, len(wcss)+1), wcss, curve="convex", direction="decreasing")
optimal_k = knee_locator.knee

print(f"Optimal number of clusters (k) suggested by KneeLocator: {optimal_k}")

# Plot with the detected elbow point
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(wcss)+1), wcss, marker='o', linestyle='--', color='b')
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal k={optimal_k}')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.legend()
plt.grid()
plt.show()
```

```python
# SChoose optimal k (based on the elbow point)
optimal_k = 3

# Step 3️⃣: Fit K-Means with Optimal k
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans_final.fit_predict(X_pca_final)  # Assign cluster labels

# Step 4️⃣: Visualize Clusters (Only works if n_components_optimal >= 2)
if X_pca_final.shape[1] >= 2:
    plt.figure(figsize=(8, 6))
    plt.scatter(X_pca_final[:, 0], X_pca_final[:, 1], c=clusters, cmap='viridis', alpha=0.7)
    plt.scatter(kmeans_final.cluster_centers_[:, 0], kmeans_final.cluster_centers_[:, 1], s=300, marker='X', c='red', label='Centroids')
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.title(f'K-Means Clustering (k={optimal_k}) on PCA Data')
    plt.legend()
    plt.grid()
    plt.show()

# Step 4️⃣: Visualize Clusters (3D Plot)
if X_pca_final.shape[1] >= 3:
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    scatter = ax.scatter(X_pca_final[:, 0], X_pca_final[:, 1], X_pca_final[:, 2], c=clusters, cmap='viridis', alpha=0.7)
    # ax.scatter(kmeans_final.cluster_centers_[:, 0], kmeans_final.cluster_centers_[:, 1], kmeans_final.cluster_centers_[:, 2], s=300, marker='X', c='red', label='Centroids')
    
    ax.set_xlabel('PCA Component 1')
    ax.set_ylabel('PCA Component 2')
    ax.set_zlabel('PCA Component 3')
    ax.set_title(f'K-Means Clustering (k={optimal_k}) in 3D PCA Space')
    ax.legend()
    plt.colorbar(scatter)
    plt.show()

# Step 5️⃣: Add Cluster Labels to DataFrame
Hormones_Enzymes['Cluster'] = clusters
```
